### Complements

> The complement of the complement is the event itself.

> The probability of a complement is $1 -$ the event itself: $P(A') = 1- P(A)$

> **eg: Probability of getting a 3 on a die is $1/6 = P(A)$**

> **The probability of not getting a 3 is $1-1/6 = 5/6 = P(A')$**

> **<font color = "yellow">A complement is the inverse of the event. In other words, a complement is an event not happening.</font>**

> ***Complement events can never occur simultaneously or at the same time. In other words, one event occurs if the other does not take place. In the example of the die, for the probability of getting 1,2,4,5,6 to occur, the probability of 3 should not occur.***

### Why do we use $P(A) + P(B) - P(A \cap B)$?

> When we want to know the chance that either event A happens, or event B happens (or both), we use this formula:
>
> $$
> P(A \text{ or } B) = P(A) + P(B) - P(A \cap B)
> $$
>
> - $P(A)$ is the probability that event A happens.
> - $P(B)$ is the probability that event B happens.
> - If we simply add $P(A)$ and $P(B)$, we accidentally count the times when both A and B happen together twice.
> - To correct this, we subtract $P(A \cap B)$, which is the probability that both A and B happen at the same time.
>
> **Simple Example:**
> Imagine you have a group of children:
> - 5 like apples (A)
> - 4 like bananas (B)
> - 2 like both apples and bananas (A and B)
>
> If you count all the apple lovers and all the banana lovers, you count the children who like both twice. So, you subtract the number who like both to get the correct total:
>
> $$
> P(A \text{ or } B) = P(A) + P(B) - P(A \cap B)
> $$
>
> This way, you only count each child once!
>
> **In summary:**
> We use this formula to avoid double-counting when events can happen at the same time.

### Additive and Multiplicative Rules of Probability

- **Additive Rule:**
  - The additive rule is used to find the probability that at least one of two events occurs (i.e., the probability of the union of two events).
  - For two mutually exclusive (disjoint) events A and B:
    $$
    P(A \text{ or } B) = P(A) + P(B)
    $$
  - For events that are not mutually exclusive:
    $$
    P(A \text{ or } B) = P(A) + P(B) - P(A \cap B)
    $$
  - **Use this rule when you want the probability that either event A or event B (or both) happens.**

- **Multiplicative Rule:**
  - The multiplicative rule is used to find the probability that two events both occur (i.e., the probability of the intersection of two events).
  - For independent events A and B:
    $$
    P(A \text{ and } B) = P(A) \times P(B)
    $$
  - For dependent events:
    $$
    P(A \text{ and } B) = P(A) \times P(B|A)
    $$
  - **Use this rule when you want the probability that both event A and event B happen together.**

We use these rules to calculate probabilities for combined events, helping us understand the likelihood of one or more events happening in various situations.

---

### Difference Between Additive and Multiplicative Rules (with Example):

- The additive rule is used for the probability of either event happening (union), while the multiplicative rule is used for the probability of both events happening together (intersection).

**Example (Additive Rule):**
Suppose you roll a die.
- Let event A be getting an even number (2, 4, 6). $P(A) = 3/6 = 0.5$
- Let event B be getting a number greater than 4 (5, 6). $P(B) = 2/6 \approx 0.33$
- Numbers that are even or greater than 4: 2, 4, 5, 6 (4 outcomes)
- $P(A \text{ or } B) = 4/6 \approx 0.67$

**Example (Multiplicative Rule - Independent Events):**
Suppose you toss a coin and roll a die.
- Probability of getting a head ($P(H)$) = $1/2$
- Probability of getting a 4 on the die ($P(4)$) = $1/6$
- The events are independent, so:
$$
P(H \text{ and } 4) = P(H) \times P(4) = 1/2 \times 1/6 = 1/12
$$

**Example (Multiplicative Rule - Dependent Events):**
Suppose in a class, 60% of students are boys ($P(B) = 0.6$). Among boys, 30% play football ($P(F|B) = 0.3$).
- Probability that a student is a boy and plays football:
$$
P(B \text{ and } F) = P(B) \times P(F|B) = 0.6 \times 0.3 = 0.18
$$

---

### Example: Probability of Being Male and Vegetarian

Suppose in a city, 60% of the population is male ($P(\text{Male}) = 0.6$). Among males, 30% are vegetarians ($P(\text{Vegetarian} \mid \text{Male}) = 0.3$).

To find the probability that a randomly selected person is both male and vegetarian:

$$
P(\text{Male and Vegetarian}) = P(\text{Male}) \times P(\text{Vegetarian} \mid \text{Male}) = 0.6 \times 0.3 = 0.18
$$

So, there is an 18% chance that a randomly selected person is both male and vegetarian.

This probability is also called the intersection of the two events: being male and being vegetarian, written as $P(\text{Male} \cap \text{Vegetarian})$.

### Mutually exclusive and independant events

> They are not the same

> `Mutually exclusive events` are events that  have no two events happening at the same time. In other words if we have one event where students play basketball and another event where students play football there cannot be an event where students play both at the same time.there is no intersection between two events or in other words the `probabilty of students playing both sports at the same time is 0`

> **`Since mutually exclusive events or sets are sets or events that don't happen at the same time that means their intersection is 0 or a null set`**

>Mutually exclusive events cannot occur simultaneously, so the probability of both events happening together is 0.

> Basically if one event occurs then the other can't

> Independent events have no influence on each other, so the probability of both events happening together is equal to the product of their individual probabilities.

> The probability of two independent events happening simultaneously or at the same time is the product of the two individual probabilities

> `Event A and Event B are mutually exclusive if P(A intersection B) = 0`

> `Event A and Event B are independant events if P(A) = P(A|B)` Applying the condition B does not affect the probability of A
or `P(B) = P(B|A)` Applying the condition A does not affect the probability of B. <u>To summarize the probability of obatining an outcome A does not affect the probability of obtaining the outcome B</u>

> **`Probability of A given B has occurred != Probability of B given A has occured`**

## Probability density function and Cummulative distributive function (continous) or known and random variables

> `Density` refers to the concentration or distribution of data points in a given range or interval. It provides information about how values are spread out within a dataset.

> **`Continous distributions are represented by PDF as the no. of outcomes we can get in a continous distribution is basically infinite`**

> In the probability formula of **`P(X) = No. of favourable outcomes / all outcomes`** 

> The probability of getting a single value in a continous distribtution is **`1/infinite which is close to 0`**

> As the denominator gets the bigger the value gets close to 0 

> So thats why the **`probability of getting a single values in a continous distribtution is 0`**

> ***Eg: Given the probability of a runner running  a mile in less than 6 mintes is equal to the probability of him running the mile in at most 6 minutes. Since time is continuous the probability of him running the mile in exactly 6 minutes is 0.***

> The PDF (Probability Density Function) itself does not give actual probabilities for specific values of a continuous random variable. Instead, it provides a relative measure of likelihood or the probability density.

>In a probability density function (PDF), if the curve peaks at a certain value (such as 400 in the below graph), `it indicates that this value has the highest relative likelihood or probability density in the distribution. The PDF curve gradually decreases on both sides of the peak, representing the decreasing likelihood of obtaining values farther away from the peak.`


> `When you calculate the cumulative distribution function (CDF) based on the PDF, the CDF starts at zero and increases as you move along the x-axis. The CDF represents the cumulative probability of obtaining values up to a given point. In other words the cumulative distribution function calulates the probability of obtaining a given value or less than that value`

> `To get the probability  of obtaining values higher than the given value we do 1-cdf`

> In the scenario you described, if the PDF curve peaks at 2000 and then the CDF curve begins to straighten out after that value, it suggests that the probability of obtaining values beyond 2000 is relatively low. As you move beyond the peak, the probabilities accumulate more slowly, indicating a lower probability density and decreasing likelihood of observing values further away from the peak.

>**<font color = "yellow">So, whether the variables are known or random, the PDF formula and the resulting plot depend on the mean and standard deviation of the distribution. These parameters provide information about the central tendency and spread of the distribution, which are crucial in determining the shape and characteristics of the PDF.**</font>

### Cummulative Distribution Function and Percentage Point Function 

> The `ppf` is the inverse of the `cdf`

> The `cdf or the cumulative distribution function calculates the cummulative probability` of a given value or less with the given mean and standard deviation

> The `ppf or the percent point function` gives the value from the cumulative distribution function or the cdf

> The cumulative distribution function (CDF) can accept either a `random variable` or a Z-score value.

> When you use the CDF function, you can provide a `random variable` as an input, and it will return the cumulative probability up to that value. For example, scipy.stats.norm.cdf(20, mean, std_dev) would give you the cumulative probability of getting a value less than or equal to 20 in a normal distribution with a given mean and standard deviation.

> Alternatively, you can provide a Z-score value to the CDF function. The Z-score represents the number of standard deviations away from the mean a specific value is. For example, scipy.stats.norm.cdf(z) would give you the cumulative probability up to the Z-score value z in a standard normal distribution.

> In summary, the CDF function in scipy.stats.norm can accept either a `random variable` or a Z-score value, depending on the context and distribution you are working with. 

> <font color = "yellow">**The default behavior for the scipy.stats.norm functions, including cdf(), pdf(), and ppf(), assumes a standard normal distribution with a mean of 0 and a standard deviation of 1**</font>.

### Difference between with and without replacement

> `"With replacement"` means the item is selected and then replaced back into the population before the next selection. the probabilities remain the same for each selection.

>> **`With Replacement`**:
If you select candies "with replacement," it means that after each selection, you put the candy back into the bag before making the next selection. In this case, the candy you selected is replaced back into the bag, and the total number of candies remains the same for subsequent selections.
For example, you reach into the bag and randomly select one candy. Let's say you get a red candy. You put it back into the bag and mix it up. Then, you reach in again and randomly select another candy. This process of selecting and replacing is called "with replacement

> `"Without replacement" `means the item is selected and not replaced, resulting in a change in probabilities for subsequent selections as the population size decreases.

>> **`Without Replacement`**:
If you select candies "without replacement," it means that after each selection, you do not put the candy back into the bag. As a result, the total number of candies decreases for subsequent selections.
For example, you reach into the bag and randomly select one candy. Let's say you get a red candy. However, this time, you don't put it back into the bag. Now, there are only 9 candies left in the bag. Then, you reach in again and randomly select another candy

## Binomial distribution

>  **`The binomial distribution is specifically designed to model situations where there are only two possible outcomes, often referred to as    "success" and "failure." These outcomes are mutually exclusive, meaning that only one of them can occur in a single trial.`**

> Keep in mind that the trial has a possibility of only `two outcomes`

> It deals with `discrete variables` which means it uses the PMF function to calculate the individiual probability and CDF to calculate the cummulative probability. 

> Prequisites for the binomial distribution
>>  1. There are two potential outcomes per trial 

>>  2. The probability of success (p) is same across all trials. eg: There is 50% probability to get heads or tails. 

>>  3. The number of trials(n) is fixed

>>  4. Each trial is independent which means the outcome of one trial doesn't affect the probability of the other trial.


> Parameters of the `binom` function
>>  1. **`n`** =  Number of trials to be conducted as per the problem.

>>  2. **`p`** =  The probability of success of said statistical outcome. 

>>  3. **`k`** =  Probability of observing `k` observations in the problem or distribution

>>  4. **`rvs`** = random variables

>>  5. **`size`** = no. of samples to be generated

## Poisson distribution

> 1. Its a discrete distribution which means there is no pdf function

> 2. It requires two parameters `k` and `mu`

> 3. The `loc` parameter allows you to shift the distribution along the x-axis. It is often used when you want to model events that start at a specific point in time or space.
 
> 4. It is assumed that the rate at which events occur are constant

> Finding outliers using a function

```python

def retrieve_outliers(df):
    threshold = 2
    mean = np.mean(df)
    std = np.std(df)
    outliers = [i for i in df if np.abs(i - mean)/std > threshold]
    return outliers
 
```

# Correlation and Covariance 



`Pearson's correlation` and `Spearman's correlation` are both statistical measures used to quantify the relationship between two variables. However, they differ in terms of the type of data they are suitable for and how they handle outliers.

**Pearson's correlation**:

>Pearson's correlation coefficient measures the strength and direction of a linear relationship between two continuous variables.

> <u>It assumes that the relationship between the variables is linear and follows a specific pattern.</u>

> It is sensitive to both the strength and direction of the relationship.

> Outliers can have a significant impact on Pearson's correlation. A single outlier with extreme values can distort the correlation coefficient and make it appear stronger or weaker than it actually is.


**Spearman's correlation**:

> Spearman's correlation coefficient measures the strength and direction of a monotonic relationship between variables.

> It does not assume a linear relationship but looks for a consistent monotonic trend (either increasing or decreasing) between the variables.

> It is based on the ranks of the data rather than the actual values.

> Spearman's correlation is less sensitive to outliers compared to Pearson's correlation. Outliers have less influence because the focus is on the order or ranks of the data rather than their actual values.

In summary, Pearson's correlation is appropriate for assessing linear relationships between continuous variables, but it is sensitive to outliers. Spearman's correlation, on the other hand, is suitable for assessing monotonic relationships and is more robust against outliers.

<font color = "yellow"> <u>A monotonic relationship means that as the values of one variable increase, the ranks of the corresponding values of the other variable also tend to increase (monotonic increasing) or decrease (monotonic decreasing).</u></font>

### Mutually exclusive statements 
> A mutually exclusive statement in statistics refers to two or more events or categories that cannot occur simultaneously. It means that if one event or category happens, the others cannot happen at the same time. In other words, these events or categories are mutually exclusive or disjoint

> For example, let's consider a simple situation where we have two categories: "heads" and "tails" for flipping a coin. If the coin is flipped and it lands on "heads," it cannot simultaneously be "tails." Similarly, if it lands on "tails," it cannot be "heads" at the same time. These two categories, "heads" and "tails," are mutually exclusive because only one of them can occur in a single coin flip

> In statistics, mutually exclusive events or categories are important because they allow us to define and analyze probabilities. If events are mutually exclusive, the probability of any one of them occurring can be calculated by summing the probabilities of each individual event.

> **`Type I Error`** (False Positive): A Type I error occurs when the null hypothesis is rejected, even though it is actually true. In other words, it's an incorrect rejection of a true null hypothesis. This error represents a false positive result, where the test concludes that there is a significant effect or relationship when, in fact, there is none. The probability of committing a Type I error is denoted by α (alpha) and is set by the researcher in advance as the significance level. This case occurs when there is lack of evidence to prove the truth of the null hypothesis. This leads to the rejection of the null hypothesis and adoption of the alternative hypothesis.

> **`Type II Error`** (False Negative): A Type II error occurs when the null hypothesis is not rejected, even though it is false. In other words, it's an incorrect acceptance of a false null hypothesis. This error represents a false negative result, where the test fails to detect a significant effect or relationship that actually exists. The probability of committing a Type II error is denoted by β (beta). This case occurs when there is evidence to prove that the null hypothesis is False but is claimed to be True. In other words the alternate hypothesis is rejected even when it is claimed to be True.

- It focuses on whether the observed data is significantly greater or significantly smaller than what is expected under the null hypothesis.

**Two-Tailed Test:**

- In a two-tailed test, we are interested in examining the effect or difference in both directions.
- It considers whether the observed data is significantly different, regardless of whether it is greater or smaller than what is expected under the null hypothesis.
- `Using the same example of the medication and test scores, a two-tailed test would determine if the scores are significantly different from what is expected, without specifying whether they are higher or lower.`

> Using the <u>`chi.cdf`</u> function we can find the `p_value`

```python

from scipy.stats import chi2

critical_value = chi2.ppf(q = 1- alpha_value, df = ddof)
chi_test_p_value = 1-chi2.cdf(x = chi_square_statistic, df = ddof)
```

```python
critical_value, chi_test_p_value
```

> <font color = "yellow">**The comparison between the chi-square test statistic and the critical value helps us determine if there is a significant deviation between the observed and expected values.**</font>

> If the **`chi-square test statistic is larger than the critical value`**, it suggests that the deviation between the observed and expected values is unlikely to be explained by random chance alone. In this case, we reject the null hypothesis and conclude that there is evidence of a significant deviation or association between the variables being studied.

> On the other hand, if **`the chi-square test statistic is smaller than the critical value`**, it suggests that the observed deviation could reasonably be attributed to random chance. In this case, we fail to reject the null hypothesis and conclude that there is not enough evidence to support a significant deviation or association between the variables.

```python
if chi_square_statistic >= critical_value:
    print("Reject null hypothesis, there is a relationship between the two variables")
else: 
    print("Retain the null hypothesis, there is no relationship between the two variables given the comparison" )


if chi_test_p_value <= alpha_value:
    print("Reject null hypothesis, there is a relationship between the two variables")
else:
    print("Retain the null hypothesis, there is no relationship between the two variables")
```
